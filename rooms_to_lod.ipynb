```python
import pandas as pd
import re
```

```python
rooms_csv = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vTRvHMQkXQdsW06bzvoOSTw0wkdB4utsbd2JwR2nzASGwkKBhQyDk0_IKu5wl8p5oaambrzr9k3hsMX/pub?gid=279546102&single=true&output=csv",dtype='str')
rooms_csv.head(2)
```

```python
roman_numerals = {'I':'1','II':'2','III':'3','IV':'4','V':'5','VI':'6','VII':'7','VIII':'8','IX':'9'}
property_ids = []
rooms_layers = []
rooms_labels = []
rooms_ids = []
for row in rooms_csv.iterrows():
    match = re.match(r"^([IVX]+?)([0-9][0-9])([0-9][0-9])(.*?)$",row[1]['Layer'])
    
    if match:
        r = roman_numerals[match.group(1)]
        i = match.group(2).replace("0","")
        p = match.group(3).replace("0","")
        room_label = match.group(4)
        property_ids.append("r{0}-i{1}-p{2}".format(r,i,p))
        rooms_layers.append(row[1]['Layer'])
        rooms_labels.append(room_label)
        rooms_ids.append("room{0}".format(row[1]['OBJECTID']))

rooms_df = pd.DataFrame(data = {"dcterms:identifier":rooms_ids,
                                "p-lod:spatially-within":property_ids,
                                "layer":rooms_layers,
                                "dcterms:title":rooms_labels})
```
